```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv(r'E:\Github\Pakistan Largest Ecommerce Dataset.csv', encoding='latin-1',  nrows = 50000, usecols=['category_name_1', 'payment_method', 
'grand_total', 'status', 'Month'])
df.shape
```

```python
fig , axes = plt.subplots(2,2, figsize=(15,10))
axes[0,0].bar(df['category_name_1'].value_counts().index, df['category_name_1'].value_counts())
axes[0,0].set_title('Category Name 1 Count')
axes[0,0].set_xlabel('Category Name 1')
axes[0,0].set_ylabel('Count')
axes[0,1].pie(df['payment_method'].value_counts(), labels=df['payment_method'].value_counts().index, autopct='%1.1f%%')
axes[0,1].set_title('Payment Method Distribution')
monthly = df['Month'].value_counts().sort_index()
axes[1,0].plot(monthly.index, monthly.values, color='black', linewidth=2, marker='o')
axes[1,0].set_title('Monthly Grand Total Count')
axes[1,0].set_xlabel('Month')
axes[1,0].set_ylabel('Count')
status_counts = df['status'].value_counts()
axes[1,1].bar(status_counts.index, status_counts.values, color='orange')
axes[1,1].set_title('Status Count')
axes[1,1].set_xlabel('Status')
axes[1,1].set_ylabel('Count')
plt.tight_layout()
plt.show()
```
