```python
# !/usr/bin/env python3
# yaha pe hum groupby ka use karenge aur data ko group karenge based on status and category_name_1.
# yaha abhi libraries import karenge aur data ko read karenge.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'E:\Github\Pakistan Largest Ecommerce Dataset.csv', encoding = 'latin-1', nrows=50000, usecols=['status', 'price', 'qty_ordered', 
                          'grand_total', 'category_name_1', 
                          'payment_method', 'Year', 'Month'])

df.shape
```

```python
# yaha pe hum groupby ka use karenge aur data ko group karenge based on status and category_name_1.
df.groupby('category_name_1')['grand_total'].sum().round(0)
```

```python
# yaha pe hum groupby ka use karenge aur data ko group karenge based on payment_method and grand_total.
df.groupby('payment_method')['grand_total'].mean()
```

```python
# yaha pe hum groupby ka use karenge aur data ko group karenge based on Month and qty_ordered.
df.groupby('Month')['qty_ordered'].count()
```

```python
# yaha pe hum groupby ka use karenge aur data ko group karenge based on category_name_1 and grand_total.
df.groupby('category_name_1').agg(
    total_orders=('grand_total', 'count'),
    avg_order=('grand_total', 'mean'),
    total_revenue=('grand_total', 'sum')
)

```

```python
# yaha pe hum groupby ka use karenge aur data ko group karenge based on category_name_1 and grand_total.
result = df.groupby('category_name_1').agg(
    total_orders=('grand_total', 'count'),
    avg_order=('grand_total', 'mean'),
    total_revenue=('grand_total', 'sum')
)

result.sort_values('total_revenue', ascending=False)
```

```python
# yaha pe hum groupby ka use karenge aur data ko group karenge based on category_name_1 and grand_total.
result_sorted = result.sort_values('total_revenue')

plt.figure(figsize=(12, 6))
plt.barh(result_sorted.index, result_sorted['total_revenue'], color='orange')
plt.title('Category Revenue Ranking')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()
```
