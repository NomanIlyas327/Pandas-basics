```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'E:\Github\Pakistan Largest Ecommerce Dataset.csv', encoding='latin-1', nrows=50000, usecols=['category_name_1', 'payment_method', 
'grand_total', 'status', 'Month'])

df.shape
```

```python
fig , axes = plt.subplots(2,2, figsize = (14,10))
```

```python
# Figure banao
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Pakistan Ecommerce Dashboard', fontsize=16)

# Chart 1 - Top 5 categories
top_cats = df['category_name_1'].value_counts().head(5)
axes[0,0].bar(top_cats.index, top_cats.values, color='steelblue')
axes[0,0].set_title('Top 5 Categories')
axes[0,0].set_xlabel('Category')
axes[0,0].set_ylabel('Orders')
axes[0,0].tick_params(axis='x', rotation=45)

# Chart 2 - Payment methods
payment = df['payment_method'].value_counts()
axes[0,1].pie(payment.values, labels=payment.index, autopct='%1.1f%%', textprops={'fontsize': 7})
axes[0,1].set_title('Payment Methods')

# Chart 3 - Monthly trend
monthly = df['Month'].value_counts().sort_index()
axes[1,0].plot(monthly.index, monthly.values, 
               color='red', linewidth=2, marker='o')
axes[1,0].set_title('Monthly Orders')
axes[1,0].set_xlabel('Month')
axes[1,0].set_ylabel('Orders')
axes[1,0].grid(True, alpha=0.3)

# Chart 4 - Order status
status = df['status'].value_counts()
axes[1,1].barh(status.index, status.values, color='orange')
axes[1,1].set_title('Order Status')
axes[1,1].set_xlabel('Count')

plt.tight_layout()
plt.show()
```
