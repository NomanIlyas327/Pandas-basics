```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "player": ["Babar", "Rizwan", "Fakhar", "Imam", "Shadab"],
    "matches": [45, 42, 38, 35, 40],
    "runs": [2100, 1890, 1650, 1420, 890],
    "sixes": [45, 32, 67, 28, 15]
}
df  = pd.DataFrame(data)
print(df)
```

```python
average_runs = df['runs'].mean()
print(average_runs)
```

```python
most_scorers = df[df['runs'] > 1500]
print("players with 1500 runs or more")
print(most_scorers)
```

```python
df['runs_per_match'] = df['runs'] / df['matches']
print(df['runs_per_match'])
print(df)
```

```python
plt.bar(most_scorers['player']  , most_scorers['runs'])
plt.title('Players vs Runs')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.tight_layout()
plt.show()
```
