```python
# !/usr/bin/env python3
# yaha par humne kuch libraries import ki hai jo data analysis aur visualization ke liye use hoti hai.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# yaha par humne ek CSV file read ki hai jisme Pakistan ke largest ecommerce dataset ka data hai. Humne sirf 50000 rows read ki hai aur kuch specific columns select kiye hai.
df = pd.read_csv('', encoding='latin-1',  nrows=50000, usecols=['status', 'price', 'qty_ordered', 'grand_total', 'category_name_1', 'payment_method', 'Year', 'Month'])
df.head()
df.shape
```

```python
# yaha par humne dataframe ke missing values check kiye hai. iske liye humne isnull() function ka use kiya hai jo dataframe ke har column me missing values ko check karta hai aur sum() function ka use karke total missing values ko count karta hai.
df.isnull().sum()
```

```python
# yaha par humne dataframe ke grand_total column ka maximum value aur mean value calculate kiya hai. iske liye humne max() aur mean() function ka use kiya hai. iske alawa humne category_name_1 column ke unique values ka count bhi calculate kiya hai aur top 5 unique values ko print kiya hai.
print(df['grand_total'].max())
print(df['grand_total'].mean())
print(df['category_name_1'].value_counts().head())
```

```python
# yaha par humne payment_method column ke unique values ka count calculate kiya hai aur unke distribution ko print kiya hai.
print(df['payment_method'].value_counts())
```

```python
# yaha par humne Year column ke unique values ka count calculate kiya hai aur unke distribution ko print kiya hai.
print(df['Year'].value_counts())
```

```python
# yaha par humne Month column ke unique values ka count calculate kiya hai aur unke distribution ko print kiya hai.
print(df['Month'].value_counts())
```

```python
# yaha par humne category_name_1 column ke unique values ke hisaab se grand_total column ka mean calculate kiya hai. iske liye humne groupby() function ka use kiya hai jo dataframe ko category_name_1 ke hisaab se group karta hai aur phir grand_total column ka mean calculate karta hai.
df.groupby('category_name_1')['grand_total'].mean()

```

```python
# yaha par humne category_name_1 column ke unique values ke hisaab se total orders ka distribution plot kiya hai. iske liye humne bar chart ka use kiya hai jo har unique value ke liye total orders ko show karta hai.
plt.bar(df['category_name_1'].value_counts().index, df['category_name_1'].value_counts().values)
plt.xticks(rotation = 45 , ha = 'right')
plt.xlabel('category_name_1')
plt.ylabel('Total Orders')
plt.title('Total Orders by Category')
plt.tight_layout()
plt.show()
```

```python
# yaha par humne payment_method column ke unique values ke hisaab se total orders ka distribution plot kiya hai. iske liye humne pie chart ka use kiya hai jo har unique value ke liye total orders ko show karta hai.
plt.pie(df['payment_method'].value_counts().values, labels=df['payment_method'].value_counts().index, autopct='%1.1f%%')
plt.title('Payment Method Distribution')
plt.tight_layout()
plt.show()
```

```python
# yaha par humne Month column ke unique values ke hisaab se total orders ka distribution plot kiya hai. iske liye humne line chart ka use kiya hai jo har unique value ke liye total orders ko show karta hai.
monthly = df['Month'].value_counts().sort_index()
plt.plot(monthly.index, monthly.values)
plt.xticks(rotation = 45 , ha = 'right')
plt.xlabel('Month')
plt.ylabel('Total Orders')
plt.title('Monthly Order Distribution')
plt.tight_layout()
plt.show()
```

```python
# yaha par humne category_name_1 column ke unique values ke hisaab se grand_total column ka mean calculate kiya hai aur uska bar chart plot kiya hai. iske liye humne groupby() function ka use kiya hai jo dataframe ko category_name_1 ke hisaab se group karta hai aur phir grand_total column ka mean calculate karta hai. phir humne sort_values() function ka use karke mean values ko ascending order me sort kiya hai aur bar chart plot kiya hai.
cat_revenue = df.groupby('category_name_1')['grand_total'].mean().sort_values()
plt.bar(cat_revenue.index, cat_revenue.values)
plt.xticks(rotation = 45 , ha = 'right')
plt.xlabel('Category Name 1')
plt.ylabel('Average Revenue')
plt.title('Average Revenue by Category')
plt.tight_layout()
plt.show()
```
