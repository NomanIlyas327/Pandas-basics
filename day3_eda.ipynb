```python
import pandas as pd
import numpy as np
```

```python
data = {"name" :  ["shirt", "Pants", "Shoes" , "Kurta" , "Jacket"],
        "price" : [1500, 2500, 3500, 1200, 4000],
        "quantity_sold" : [30, 15, 20, 45, 10]}

df = pd.DataFrame(data)
print(df)
```

```python
mehanga_product = df[df['price'] == df['price'].max()]
print("Mehanga Product:")
print(mehanga_product)
```

```python
total_quantity = df['quantity_sold'].sum()
print("Total Quantity Sold:", total_quantity )
```

```python
df['total_revenue'] = df['price'] * df['quantity_sold']
print(df['total_revenue'])
```

```python
print(df)
```

```python
filtered_product = df[df['price'] > 2000]
print("Products with price greater than 2000:")
print(filtered_product)
```

```python
most_revenue_product = df[df['total_revenue'] == df['total_revenue'].max()]
print("Product with the most revenue:")
print(most_revenue_product)
```

```python
import matplotlib.pyplot as plt
plt.bar(df['name'] , df['total_revenue'])
plt.title("Total Revenue by Product")
plt.xlabel("name")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
